<h1 style="color:red;font-size:120%">Only for relatively small query sequences, use the below cell's query method</h1>

 <span style="text-decoration:underline">ChatGPT Explanation:</span> <br>
 
1. Loading the Query File:
    * We load the query sequences (van_bakta_hsic_seq.fa) into memory because this is typically a smaller file compared to the target.
    * query_sequences is a list of the sequences from the query file.

2. Iterate Through the Target File:
    * The target file (HC_VRE0008_assembly_complete.ffn) is read one sequence at a time using SeqIO.parse(). This is memory-efficient because we only keep one sequence in memory at a time.

3. Check for Subsequence Matches:
    * For each target sequence, we iterate through all the query sequences and check if any of them are present (query_seq in target_seq).
    * If a match is found, we write the entire target sequence to an output file (matching_sequences.fasta). This output is optional, but it can be useful for further analysis.

4. Output File:
    * Matching sequences from the target file are saved to matching_sequences.fasta.

___Optimizations for Memory Efficiency:___
* Query Sequences Loaded Once: We load van_bakta_hsic_seq.fa once because it’s assumed to be smaller. If it’s also large, consider iterating through it in chunks or using hashing.
* Target File Processed Sequentially: HC_VRE0008_assembly_complete.ffn is processed one sequence at a time, keeping memory usage low.
* Stop at First Match for a Target: If a query sequence matches a target sequence, the loop breaks, saving computation time.<br>

This approach is efficient in terms of memory and scales reasonably well with larger files, assuming the query set (van_bakta_hsic_seq.fa) is not excessively large compared to the target set. If both files are huge, further optimizations like hashing sequences and using more efficient data structures for search (e.g., suffix trees) might be required.

````python
from Bio import SeqIO
import pandas as pd

def find_matching_and_non_matching_sequences(query_file, target_file, output_file, report_file):
    """
    Find all sequences in the query_file that are present in the target_file,
    save the matching sequences from the target file to an output file, and
    save both matching and non-matching sequences report to a text file and DataFrame.

    Parameters:
        query_file (str): Path to the query multi-FASTA file.
        target_file (str): Path to the target multi-FASTA file.
        output_file (str): Path to save the matching sequences in FASTA format.
        report_file (str): Path to save the report of matching and non-matching sequences.
    """
    # Load query sequences into a dictionary (keyed by description)
    query_sequences = {record.description: str(record.seq) for record in SeqIO.parse(query_file, "fasta")}

    matched_records = []  # List to hold matching target records
    matching_info = []    # List to hold matching sequences info (query and target labels)
    non_matching_query = set(query_sequences.keys())  # Set to track non-matching query sequences

    # Iterate through each sequence in the target file
    for target_record in SeqIO.parse(target_file, "fasta"):
        target_seq = str(target_record.seq)
        target_id = target_record.description
        match_found = False

        # Check if any of the query sequences are present in the target sequence
        for query_id, query_seq in query_sequences.items():
            if query_seq in target_seq:
                matched_records.append(target_record)
                matching_info.append((query_id, target_id))
                if query_id in non_matching_query:
                    non_matching_query.remove(query_id)  # Remove from non-matching set
                match_found = True
                break  # Stop after finding the first match for efficiency

    # Write matching target records to the output file
    with open(output_file, "w") as out_handle:
        SeqIO.write(matched_records, out_handle, "fasta")

    # Prepare lists for DataFrame
    query_list = []
    target_list = []

    # Add matching sequences to DataFrame lists
    for query_id, target_id in matching_info:
        query_list.append(query_id)
        target_list.append(target_id)

    # Add non-matching queries to DataFrame lists (Target is None)
    for query_id in non_matching_query:
        query_list.append(query_id)
        target_list.append(None)

    # Create a DataFrame with the matching and non-matching information
    df = pd.DataFrame({'Query': query_list, 'Target': target_list})

    # Write matching and non-matching sequences report to a text file and print it to the console
    with open(report_file, "w") as report_handle:
        report_handle.write("Matching Sequences:\n")
        print("# Matching Sequences:")
        for query_id, target_id in matching_info:
            line = f"Query: {query_id} <-> Target: {target_id}\n"
            report_handle.write(line)
            print(line.strip())

        report_handle.write("\n# Non-Matching Query Sequences:\n")
        print("\nNon-Matching Query Sequences:")
        for query_id in non_matching_query:
            line = f"Query: {query_id}\n"
            report_handle.write(line)
            print(line.strip())

    # Print final message
    print(f"\nMatching sequences have been saved to {output_file}")
    print(f"Report has been saved to {report_file}")
    print("Dataframe is returned, so must save it in a variable")

    return df
````
