```python
from Bio import SeqIO 
from Bio.SeqRecord import SeqRecord

# Parse the original FASTA file
fa = SeqIO.parse("/home/jee/bakta/SRR647_Feb2023/bakta_roary_output_1677454064/pan_genome_reference.fa", "fasta")

dox_HSIC_DE_seq = []

# Iterate through records and filter based on the gene name
for record in fa:
    # The gene name would be "description"
    if record.description.split(' ', 1)[-1] in dox_HSIC_DE_ls:
        dox_HSIC_DE_seq.append(record)

display(dox_HSIC_DE_seq)

# Write the filtered records into a new multi-FASTA file
output_file = "/home/jee/rna_seq/bowtie2/output_sam/dox_HSIC_no_DE_seq.fa"
SeqIO.write(dox_HSIC_DE_seq, output_file, "fasta")

print(f"Multi-FASTA file saved to {output_file}")
```
